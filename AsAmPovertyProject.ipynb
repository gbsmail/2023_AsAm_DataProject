```python
## drop all the change comparison sections and the percentage columns
dfa = df.drop(['BCT2020','NTA Type','Pop_10P','HHPop_10P', 'GQ_10P', 'InstGQ_10P', 'AvHHSz_10P',
       'PopU18_10', 'Hsp_10P', 'WNH_10P', 'BNH_10P', 'ANH_10P','ONH_10P', 'NH2pl_10P', 'HUnits_10P', 
       'OcHU_10P', 'VacHU_10P','Pop_20P',
       'HHPop_20P','GQ_20P', 'InstGQ_20P',
       'AvHHSz_20P', 'PopU18_20P', 'Hsp_20P',
       'WNH_20P', 'BNH_20P', 'ANH_20P',
       'ONH_20P', 'NH2pl_20P', 'HUnits_20P',
     'OcHU_20P', 'VacHU_20P','Pop_Ch', 'Pop_PCh',
       'HHPop_Ch', 'HHPop_PCh', 'GQ_Ch', 'GQ_PCh', 'InstGQ_Ch', 'InstGQ_PCh',
       'AvHHSz_Ch', 'AvHHSz_PCh', 'PopU18_Ch', 'PopU18_PCh', 'Hsp_Ch',
       'Hsp_PCh', 'WNH_Ch', 'WNH_PCh', 'BNH_Ch', 'BNH_PCh', 'ANH_Ch',
       'ANH_PCh', 'ONH_Ch', 'ONH_PCh', 'NH2pl_Ch', 'NH2pl_PCh', 'HUnits_Ch',
       'HUnits_PCh', 'OcHU_Ch', 'OcHU_PCh', 'VacHU_Ch', 'VacHU_PCh'], axis=1)
dfa
```

```python
## keep GeoID to turn into Community District # 
dfb = dfa[["GeoType", "Borough", "GeoID", "Name","ANH_10","ANH_20"]].copy()
dfb
```

```python
## compare Asian American population 2010 vs 2020 
df[list(dfb.columns)[4:6]]
```

```python
## convert to floating point numbers 
## "COERCE" means turning non numbers into a NAN value 
dfb[list(dfb.columns)[4:6]] = df[list(dfb.columns)[4:6]].apply(pd.to_numeric, errors = "coerce")
dfb.info()
```

```python
## rename columns 
## can put the same variable in to save the changes back into the variable
dfb = dfb.rename(columns={"GeoID": "Community District", "ANH_10": "2010 Asian Population", "ANH_20": "2020 Asian Population"})
dfb
```

```python
## drop the first number in the CD marker to match the snap recipients data, below 
dfb['Community District'] = dfb['Community District'].astype(str).str.slice(start=1)
dfb
```

```python
dfb.to_csv("asam-test-export.csv", index = False, encoding = "UTF8")
```
