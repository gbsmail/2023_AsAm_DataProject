```python
df1 = pd.merge(dfc,dfs)
df1
```

```python
df1["Asian"] = pd.to_numeric(df1["Asian"], errors = "coerce")
df1.info()
```

```python
df = df1[['Year', 'Borough','updated_CDN', 'Asian', 'BC_SNAP_Recipients', 'BC_SNAP_Households', 'Average household size',
          'Total housing units', 'Occupied housing units', 'Vacant housing units', 'Month']]
df
```
